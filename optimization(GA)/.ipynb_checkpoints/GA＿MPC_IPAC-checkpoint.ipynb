```python
import torch
import torch.nn as nn
import torch.nn.functional as F
class softsensor(nn.Module):
    def __init__(self,input_feature_dim,hidden_feature_dim1,hidden_feature_dim2,classes_num,gpu):
        super(softsensor, self).__init__()
        
        self.input_feature_dim=input_feature_dim
        self.hidden_feature_dim1=hidden_feature_dim1
        self.hidden_feature_dim2=hidden_feature_dim2
#
        #初始化LSTM       
        self.lstm1=nn.LSTMCell(input_feature_dim,hidden_feature_dim1)   
        self.lstm2=nn.LSTMCell(hidden_feature_dim1,hidden_feature_dim2) 
        self.linear1=nn.Linear(hidden_feature_dim2,classes_num)
        
    def init_hidden(self,batch_size,gpu):
        if gpu==True:
            h01=torch.zeros(batch_size,self.hidden_feature_dim1).cuda()
            c01=torch.zeros(batch_size,self.hidden_feature_dim1).cuda()
            h02=torch.zeros(batch_size,self.hidden_feature_dim2).cuda()
            c02=torch.zeros(batch_size,self.hidden_feature_dim2).cuda()
        else:
            h01=torch.zeros(batch_size,self.hidden_feature_dim1)
            c01=torch.zeros(batch_size,self.hidden_feature_dim1)
            h02=torch.zeros(batch_size,self.hidden_feature_dim2)
            c02=torch.zeros(batch_size,self.hidden_feature_dim2)
        
        return (h01,c01,h02,c02)
    '''
    定義正向傳播
    ''' 
    def forward(self,input,batchsize,gpu):
        h01,c01,h02,c02=self.init_hidden(batchsize,gpu)
        h1=h01
        c1=c01
        h2=h02
        c2=c02
        output = []
        for i in range(input.size()[1]):
            xt=input[:,i,:]
            h1,c1=self.lstm1(xt.float(),(h01,c01))
            h1=F.tanh(h1)
            #output=F.dropout(output, p=0.1,training=self.training)
            h2,c2=self.lstm2(h1,(h2,c2))
            h2=F.tanh(h2)
            #output=F.dropout(output, p=0.1,training=self.training)
            output.append(h2)
        output=torch.stack(output)
        output=output.permute(1, 0, 2)
        linear_output=self.linear1(output[:,10:])
        return linear_output
    def inference(self,input,batchsize,gpu,past):
        h01,c01,h02,c02=self.init_hidden(batchsize,gpu)
        h1=h01
        c1=c01
        h2=h02
        c2=c02
        outputs = []
        y=[]
        for i in range(input.size()[1]):
            if i<=(past-1):
                xt=input[:,i,:]      
                h1,c1=self.lstm1(xt.float(),(h01,c01))
                h1=F.tanh(h1)
                h2,c2=self.lstm2(h1,(h2,c2))
                h2=F.tanh(h2)
                output=self.linear1(h2)
                outputs.append(output)
                y=output
            if i>(past-1):
                xt=input[:,i,:]
                xt[0,3]=y[0][0]
                xt[0,4]=y[0][1]
                h1,c1=self.lstm1(xt.float(),(h01,c01))
                h1=F.tanh(h1)
                h2,c2=self.lstm2(h1,(h2,c2))
                h2=F.tanh(h2)
                output=self.linear1(h2)
                outputs.append(output)
                y=output
            
        outputs=torch.stack(outputs)
        return outputs
    
    def predict_test(self,input,batchsize,gpu):
        h01,c01,h02,c02=self.init_hidden(batchsize,gpu)
        h1=h01
        c1=c01
        h2=h02
        c2=c02
        outputs = []
        for i in range(input.size()[1]):
            xt=input[:,i,:]      
            h1,c1=self.lstm1(xt.float(),(h01,c01))
            h1=F.tanh(h1)
            h2,c2=self.lstm2(h1,(h2,c2))
            h2=F.tanh(h2)
            output=self.linear1(h2)
            outputs.append(output)
        print(torch.stack(outputs).view(input.size()[1],2).detach().numpy())

model = softsensor(5,32,16,2,gpu=False)
model.load_state_dict(torch.load("lstmcell_32_16_m2o_win30_0615.pt", map_location={'cuda:0': 'cpu'}), strict=False)
print(model)
```

```python
import numpy as np
from scipy.optimize import fsolve, basinhopping
import random
import time
np.set_printoptions(threshold=np.inf)
 
# 根據解的精度確定染色體(chromosome)的長度
# 需要根據決策變數的上下邊界來確定
def getEncodedLength(delta=[], boundarylist=[]):
    # 每個變數的編碼長度
    lengths = []
    for i, element in enumerate(boundarylist):
        lower = element[0]
        upper = element[1]
        # lamnda 代表匿名函數f(x)=0,50代表搜索的初始解
        res = fsolve(lambda x: ((upper - lower) * 1 / delta[i]) - (2 ** x - 1), 50)
        length = int(np.floor(res[0]))
        lengths.append(length)
    return lengths
    pass
 
 
# 隨機生成初始編碼種群
def getIntialPopulation(encodelength, populationSize):
    # 隨機化初始種群為0
    chromosomes = np.zeros((populationSize, sum(encodelength)), dtype=np.uint8)
    for i in range(populationSize):
        chromosomes[i, :] = np.random.randint(0, 2, sum(encodelength))
    #print('chromosomes shape:', chromosomes.shape)
    return chromosomes
 
 
# 染色體解碼得到表現型的解
def decodedChromosome(encodelength, chromosomes, boundarylist, delta=0.0001):
    populations = chromosomes.shape[0]
    variables = len(encodelength)
    #print('encodelength:',encodelength)
    #print('variables:',variables)
    decodedvalues = np.zeros((populations, variables))
   # print(decodedvalues.shape)
    for k, chromosome in enumerate(chromosomes):
        chromosome = chromosome.tolist()
        start = 0
        for index, length in enumerate(encodelength):
            # 將一個染色體進行拆分，得到染色體片段
            power = length - 1
            # 解碼得到的10進制數字
            demical = 0
            #print(start)
            #print(length+start)
            #print('chromosom:',chromosome[start:length + start])
            for i in range(start, length + start):
                demical += chromosome[i] * (2 ** power)
                power -= 1
            lower = boundarylist[index][0]
            upper = boundarylist[index][1]
            decodedvalue = lower + demical * (upper - lower) / (2 ** length - 1)
            #print('decodevalue:',decodedvalue)
            decodedvalues[k, index] = decodedvalue
            # 開始去下一段染色體的編碼
            start = length + start
    #print('decodevalues:',decodedvalues)
    return decodedvalues
 
 
# 得到個體的適應度值及每個個體被選擇的累積概率
def getFitnessValue(func, chromosomesdecoded):
    # 得到種群規模和決策變數的個數
    population, nums = chromosomesdecoded.shape
    # 初始化種群的適應度值為0
    fitnessvalues = np.zeros((population, 1))
    # 計算適應度值
    for i in range(population):
        #print(func(chromosomesdecoded[i, :]))
        fitnessvalues[i, 0] = func(chromosomesdecoded[i, :])
        
    # 計算每個染色體被選擇的概率
    probability = fitnessvalues / np.sum(fitnessvalues)
    # 得到每個染色體被選中的累積概率
    cum_probability = np.cumsum(probability)
    return fitnessvalues, cum_probability
 
 
# 新種群選擇
def selectNewPopulation(chromosomes, cum_probability):
    m, n = chromosomes.shape
    newpopulation = np.zeros((m, n), dtype=np.uint8)
    # 隨機產生M個概率值
    randoms = np.random.rand(m)
    for i, randoma in enumerate(randoms):
        logical = cum_probability >= randoma
        index = np.where(logical == 1)
        # index是tuple,tuple中元素是ndarray
        newpopulation[i, :] = chromosomes[index[0][0], :]
    return newpopulation
    pass
 
 
# 新種群交叉
def crossover(population, Pc):
    """
    :param population: 新種群
    :param Pc: 交叉概率默認是0.8
    :return: 交叉後得到的新種群
    """
    # 根據交叉概率計算需要進行交叉的個體個數
    m, n = population.shape
    numbers = np.uint8(m * Pc)
    # 確保進行交叉的染色體個數是偶數個
    if numbers % 2 != 0:
        numbers += 1
    # 交叉後得到的新種群
    updatepopulation = np.zeros((m, n), dtype=np.uint8)
    # 產生隨機索引
    index = random.sample(range(m), numbers)
    # 不進行交叉的染色體進行複製
    for i in range(m):
        if not index.__contains__(i):
            updatepopulation[i, :] = population[i, :]
    # crossover
    while len(index) > 0:
        a = index.pop()
        b = index.pop()
        # 隨機產生一個交叉點
        crossoverPoint = random.sample(range(1, n), 1)
        crossoverPoint = crossoverPoint[0]
        # one-single-point crossover
        updatepopulation[a, 0:crossoverPoint] = population[a, 0:crossoverPoint]
        updatepopulation[a, crossoverPoint:] = population[b, crossoverPoint:]
        updatepopulation[b, 0:crossoverPoint] = population[b, 0:crossoverPoint]
        updatepopulation[b, crossoverPoint:] = population[a, crossoverPoint:]
    return updatepopulation
    pass
 
 
# 染色體變異
def mutation(population, Pm):
    """
    :param population: 經交叉後得到的種群
    :param Pm: 變異概率默認是0.01
    :return: 經變異操作後的新種群
    """
    updatepopulation = np.copy(population)
    m, n = population.shape
    # 計算需要變異的基因個數
    gene_num = np.uint8(m * n * Pm)
    # 將所有的基因按照序號進行10進制編碼，則共有m*n個基因
    # 隨機抽取gene_num個基因進行基本位變異
    mutationGeneIndex = random.sample(range(0, m * n), gene_num)
    # 確定每個將要變異的基因在整個染色體中的基因座(即基因的具體位置)
    for gene in mutationGeneIndex:
        # 確定變異基因位於第幾個染色體
        chromosomeIndex = gene // n
        # 確定變異基因位於當前染色體的第幾個基因位
        geneIndex = gene % n
        # mutation
        if updatepopulation[chromosomeIndex, geneIndex] == 0:
            updatepopulation[chromosomeIndex, geneIndex] = 1
        else:
            updatepopulation[chromosomeIndex, geneIndex] = 0
    return updatepopulation
    pass
 
 
# 定義適應度函數
def fitnessFunction():
    return lambda x: sum_mse(x)
    pass
def sum_mse(x):
    raw_set=pd.read_csv('rawdata.csv')
    raw_set_norm=(raw_set-data_describe.loc["mean"])/data_describe.loc["std"]
    mse_sum=0
    p=model_predict(x)
    for i in range(21):
        mse=(p[i][0]-raw_set_norm.loc[28,'IPAC'])**2+(p[i][1]-(raw_set_norm.loc[28,'WATER']))**2
        #mse=(p[i][0]-0.19160000000000288)**2+(p[i][1]-3.522000000000025)**2
        mse_sum+=mse
        
    delta_u_sum=0
    delta_u=[]
    for i in range(int((len(x)/2)-1)):
        delta_u_B8=(x[2*i+2]-x[2*i])**2
        delta_u_B10=(x[2*i+3]-x[2*i+1])**2
        delta_u_sum=delta_u_sum+delta_u_B8+delta_u_B10
        delta_u.append([delta_u_B8,delta_u_B10])
    P=0
    for j in delta_u:
        if j[0]>0.3 or j[1]>0.3:
        #0.32 0.7
            P=1000
        else:
            P=0
    ob=mse_sum+0.5*delta_u_sum+P
    return 1/(ob)
    

 
def main(N_GENERATIONS,decisionVariables,delta,CROSS_RATE,MUTATION_RATE,POP_SIZE):    
    # 每次反覆運算得到的最優解
    optimalSolutions = []
    optimalValues = []
    # 決策變數的取值範圍
    decisionVariables = decisionVariables
    delta=delta
    # 得到染色體編碼長度
    lengthEncode = getEncodedLength(delta=delta,boundarylist=decisionVariables)
    #print(lengthEncode)
    print("---------------------開始優化------------------------")
    for iteration in range(N_GENERATIONS):
        #print("-------------------------------------------------------------------")
        #print("第{}代開始繁衍".format(iteration+1))
        # 得到初始種群編碼
        chromosomesEncoded = getIntialPopulation(lengthEncode, POP_SIZE)
        # 種群解碼
        decoded = decodedChromosome(lengthEncode, chromosomesEncoded, decisionVariables)
        #print(decoded.shape)
        # 得到個體適應度值和個體的累積概率
        evalvalues, cum_proba = getFitnessValue(fitnessFunction(), decoded)
        # 選擇新的種群
        newpopulations = selectNewPopulation(chromosomesEncoded, cum_proba)
        # 進行交叉操作
        crossoverpopulation = crossover(newpopulations,CROSS_RATE)
        # mutation
        mutationpopulation = mutation(crossoverpopulation,MUTATION_RATE)
        # 將變異後的種群解碼，得到每輪反覆運算最終的種群
        final_decoded = decodedChromosome(lengthEncode, mutationpopulation, decisionVariables)
        #print(final_decoded)
        # 適應度評價
        fitnessvalues, cum_individual_proba = getFitnessValue(fitnessFunction(), final_decoded)
        # 搜索每次反覆運算的最優解，以及最優解對應的目標函數的取值
        optimalValues.append(np.max(list(fitnessvalues)))
        index = np.where(fitnessvalues == max(list(fitnessvalues)))
        optimalSolutions.append(final_decoded[index[0][0], :])
    # 搜索最優解
    optimalValue = np.max(optimalValues)
    optimalIndex = np.where(optimalValues == optimalValue)
    optimalSolution = optimalSolutions[optimalIndex[0][0]]
    return optimalSolution, optimalValue

```

```python
def MPC_test():
    
    tStart = time.time()
    x_num=8
    POP_SIZE = 40  # population size
    CROSS_RATE = 0.75        # mating probability (DNA crossover)
    MUTATION_RATE = 0.5   # mutation probability
    N_GENERATIONS = 10
    
    raw_set=pd.read_csv('rawdata.csv')
    raw_set_norm=(raw_set-data_describe.loc["mean"])/data_describe.loc["std"]
    print("----------------優化目標----------------------------")
    #print("IPAC實際值 :  ",raw_set.loc[58,'IPAC'])
    #print("IPAC norm :  ",raw_set_norm.loc[58,'IPAC'])
    #print("WATER實際值 :  ",raw_set.loc[58,'WATER'])
    #print("WATER norm :  ",raw_set_norm.loc[58,'WATER'])
    
    decisionVariables = make_decisionVariables(data_describe)
    delta=make_delta(data_describe)

    solution, value = main(N_GENERATIONS,decisionVariables,delta,CROSS_RATE,MUTATION_RATE,POP_SIZE)
    

    tEnd = time.time()#計時結束
    print ("計算時間: {} s".format(round(tEnd - tStart),1))
    print('最優目标函数值:', (1/value))

    #MV 處理
    
    X=pd.DataFrame(columns=["B8.SP","B10.SP"])
    for i in range(4):
        X=X.append({"B8.SP": solution[2*i],"B10.SP":solution[2*i+1]},ignore_index=True)
    X["B8.SP"]=X["B8.SP"]*data_describe.loc["std","B8.SP"]+data_describe.loc["mean","B8.SP"]
    X["B10.SP"]=X["B10.SP"]*data_describe.loc["std","B10.SP"]+data_describe.loc["mean","B10.SP"]
    print("MV:")
    print(X.loc[0])
    mpcPath=pd.read_csv("mpcPath.csv")
    mpcPath.loc[mpcPath.index[-1],"B8.SP"]=X["B8.SP"][0]
    mpcPath.loc[mpcPath.index[-1],"B10.SP"]=X["B10.SP"][0]
    
    print("------------------------mpc每步優化y過程----------------------------")
    rawdata=pd.read_csv("pytorch_mpc_testdata_forMPCtest.csv")
    a=model_predict_check(solution)
    y=pd.DataFrame(columns=["IPAC","WATER"])
    for i in range(len(a)): 
        y=y.append({'IPAC': a[i][0],'WATER': a[i][1]},ignore_index=True)
    y["IPAC"]=y["IPAC"]*data_describe.loc["std","IPAC"]+data_describe.loc["mean","IPAC"]
    y['WATER']=y['WATER']*data_describe.loc["std",'WATER']+data_describe.loc["mean","WATER"]
    data_plot(y,rawdata)
    '''
    print("y 預測值第59個時間點:")
    print(y.loc[58])
    print("y 實際值第60個時間點:")
    print(rawdata.loc[59,["IPAC","WATER"]])
    '''
    
    
    print("------------------------mpc回饋後每步y----------------------------")
    new_mpcPath=mpcPath
    newrow=mpcPath.iloc[-1]
    new_mpcPath=new_mpcPath.append(newrow)
    #print(new_mpcPath[8:])
    new_mpcPath.to_csv("mpcPath.csv",index=False)
    
    plt.figure(figsize=(10,12))
    plt.subplot(4,1,3)
    plt.title("IPAC")
    plt.scatter(mpcPath["IPAC"].index,mpcPath["IPAC"], color='r',s=5)
    plt.subplot(4,1,4)
    plt.title("WATER")
    plt.scatter(mpcPath["WATER"].index,mpcPath["WATER"], color='b',s=5)
    plt.subplot(4,1,1)
    plt.title("B8.SP")
    plt.scatter(mpcPath["B8.SP"].index,mpcPath["B8.SP"], color='g',s=5)
    plt.subplot(4,1,2)
    plt.title("B10.SP")
    plt.scatter(mpcPath["B10.SP"].index,mpcPath["B10.SP"], color='k',s=3)
    
    plt.tight_layout()
    plt.show()
    print('最優目标函数值:', (1/value))
    print("MV:")
    print(X.loc[0])
```
